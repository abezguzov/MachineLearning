```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (20,10)
from mpl_toolkits.mplot3d import Axes3D
from math import sqrt
```

```python
data = pd.read_csv("student.csv")
```

```python
data.head()
```

```python
math = data.Math.values
read = data.Reading.values
write = data.Writing.values
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(math, read, write, color='#ef1234')
ax.set_xlabel("Math scores", fontsize=20)
ax.set_ylabel("Reading scores", fontsize=20)
ax.set_zlabel("Writing scores", fontsize=20)
plt.show()
```

```python
m = len(math)
```

```python
x_zero = np.ones(m)
X = np.array([x_zero, math, read]).T
```

```python
W = np.array([0,0,0])
```

```python
Y = np.array(write)
```

```python
alpha = 0.0001
```

```python
def cost_function(X, Y, W):
    m = len(Y)
    J = sum((X.dot(W) - Y) ** 2)/(2 * m)
    return J
```

```python
initial_cost = cost_function(X,Y,W)
```

```python
initial_cost
```

```python
def gradient_descent(X, Y, W, alpha, iterations):
    grad_change = []
    cost_history = [0] * iterations
    for iteration in range(iterations):
        h = X.dot(W)
        loss = h - Y
        gradient = X.T.dot(loss) / m
        grad_change.append(list(W))
        W = W - alpha*gradient
        cost = cost_function(X,Y,W)
        cost_history[iteration] = cost
    return W, cost_history, grad_change
```

```python
new_weights, cost_history, grad_change = gradient_descent(X,Y,W,alpha,1000)
```

```python
def rmse(Y, Y_pred):
    rmse = sqrt(sum((Y_pred - Y) ** 2)/len(Y))
    return rmse
```

```python
def r2score(Y, Y_pred):
    mean_y = np.mean(Y)
    ss_tot = sum((Y - mean_y) ** 2)
    ss_res = sum((Y - Y_pred) ** 2)
    r2 = 1 - (ss_res / ss_tot)
    return r2
```

```python
Y_pred = X.dot(new_weights)
```

```python
rmse(Y, Y_pred)
```

```python
cost_function(X,Y,new_weights)
```

```python
r2score(Y, Y_pred)
```

### Scikit-Learn Approach

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = np.array([math, read]).T
Y = np.array(write)
```

```python
reg = LinearRegression()
reg = reg.fit(X,Y)
Y_pred = reg.predict(X)
```

```python
rmse = np.sqrt(mean_squared_error(Y_pred, Y))
```

```python
r2 = reg.score(X,Y)
```
